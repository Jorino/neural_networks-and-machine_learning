```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-6,6,10000)
y = 1 / (1 + np.exp(-x))
```

```python
def relu(k):
    return np.maximum(0,k)
```

```python
j = np.linspace(-6,6,10000)
k = relu(j)

```

```python
# https://matplotlib.org/3.1.1/gallery/subplots_axes_and_figures/figure_title.html
# https://stackoverflow.com/questions/42818361/how-to-make-two-plots-side-by-side-using-python

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axs[0].plot(x, y, "b")
axs[0].set_title("Logistinen funktio")
axs[1].plot(j, k, "r")
axs[1].set_title("Askelfunktio")
fig.tight_layout()
```
